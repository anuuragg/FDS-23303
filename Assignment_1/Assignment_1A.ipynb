```python
import pandas as pd

pd.set_option('display.width', None)
pd.set_option('display.max_columns', None)
rawData = pd.read_csv("D:/Fundamentals of DS - BTECCE23303/Assignment_1/dataset1.csv")
rawData
```

```python
structuredData = rawData
```

```python
structuredData["Price"] = structuredData["Price"].astype(str)
structuredData["Price"] = structuredData["Price"].str.replace('$', '')
structuredData["Price"] = structuredData["Price"].astype(float)
data_mean = structuredData["Price"].mean().__round__(2)
data_median = structuredData["Price"].median()
print("Mean: " + str(data_mean))
print("Median: " + str(data_median))
```

```python
data_max = structuredData["Price"].max()
data_min = structuredData["Price"].min()
data_range = data_max - data_min

print("Max: ", data_max)
print("Min: ", data_min)
print("Range: ", data_range)
```
