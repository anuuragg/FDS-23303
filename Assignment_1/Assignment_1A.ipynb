#### B. Perform data pre-processing tasks such as data cleaning, handling missing values, and data transformation.

Handled the missing values, by replacing them my the ```mode``` of that column:

````python
for column in Cars_data.columns:
    mode_value = Cars_data[column].mode()[0]
    Cars_data[column] = Cars_data[column].fillna(mode_value)
Cars_data
````
